```python
from qiskit import*
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
qr = QuantumRegister(2) #build a 2 qubit Quantum Register
cr = ClassicalRegister(2) #build a 2 bit Classical Register to store values of measurements

circuit = QuantumCircuit(qr,cr) #create circuit with the two registers
```

```python
circuit.draw()
```

```python
circuit.h(qr[0]) #apply a Hadamard gate on the first qubit to create entaglement
circuit.draw(output='mpl') #prettier plot
```

```python
circuit.cx(qr[0], qr[1]) #create a CNOT gate, with the first qubit as control and the second as the target
circuit.draw(output='mpl')
```

```python
circuit.measure(qr,cr) #measure both qubits
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator') #simulate the circuit 
result=execute(circuit, backend = simulator).result()
```

```python
plot_histogram(result.get_counts(circuit))
```
